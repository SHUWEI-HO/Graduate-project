```python
from dtaidistance.subsequence.dtw import subsequence_alignment
import numpy as np
import pandas as pd
import pickle 


s1 = np.load('Dataset\Points\HalfSquat\Points\HalfSquat-1.npy')
s2 = np.load('Dataset\Points\HalfSquat\Points\HalfSquat-2.npy')
Refer = np.load('Dataset\Points\HalfSquat\Points\HalfSquat-3.npy')

total1 = []
seq = []
for s in s1:
    for p in s:
        seq.append(p[0])
    total1.append(seq)
    seq=[]

total2 = []
seq = []
for s in s2:
    for p in s:
        seq.append(p[0])
    total2.append(seq)
    seq = []


# Input = np.concatenate((total1, total2), axis=0)

# SA = subsequence_alignment(Input, Refer)
```

```python
p1, p2, p3 = [], [], []
for r in total1:
    p1.append(r[6])
    p2.append(r[8])
    p3.append(r[10])


t1, t2, t3 = [], [], []
for r in total1:
    t1.append(r[6])
    t2.append(r[8])
    t3.append(r[10])
```

```python
from dtaidistance import dtw_visualisation as dtwvis

refer1, refer2, refer3 = [], [], []
for ra in Refer:
    refer1.append(ra[6][0])
    refer2.append(ra[8][0])
    refer3.append(ra[10][0])

input1 = np.concatenate((p1, t1), axis=0)
input2 = np.concatenate((p2, t2), axis=0)
input3 = np.concatenate((p3, t3), axis=0)


SA1 = subsequence_alignment(refer1, input1)
SA2 = subsequence_alignment(input2, refer2)
SA3 = subsequence_alignment(input3, refer3)
fig, ax = dtwvis.plot_warpingpaths(input1, refer1, SA1.warping_paths(), path=-1)
for kmatch in SA1.kbest_matches(3):
  print(kmatch.value)
  print(kmatch.path)
  print(kmatch.segment)
  dtwvis.plot_warpingpaths_addpath(ax, kmatch.path)
```
